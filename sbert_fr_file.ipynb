```python
bs=df['job details'].iloc[11]
BSH=bs.replace("\\xa0"," ")
BSH
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')

#encode the sentences
embeddings1 = model.encode(sentences_meta, convert_to_tensor=True)
embeddings2= model.encode(sentences_rena, convert_to_tensor=True)
embeddings3= model.encode(sentences_BSH, convert_to_tensor=True)
print(embeddings1)

#compute the similarity scores
cosine_scores = util.cos_sim(embeddings1, embeddings2)
cosine_scores2= util.cos_sim(embeddings1, embeddings3)

print(cosine_scores)

print(cosine_scores2)
```

```python
print(type(cosine_scores2))
```

```python
print(embeddings1.size())
print(embeddings2.size())
print(embeddings3.size())

print(cosine_scores.size())
print(cosine_scores2.size())
```

```python
import torch

# Select top 13 cosine similarity scores
k = 13
top_k_scores, top_k_indices = torch.topk(cosine_scores.view(-1), k=k)
top_k_scores2, top_k_indices2 = torch.topk(cosine_scores2.view(-1), k=k)

print(top_k_scores)
print(top_k_indices)

print(top_k_scores2)
print(top_k_indices2)
```

```python
average_score = torch.mean(top_k_scores)
average_score2 = torch.mean(top_k_scores2)

print("Average cosine similarity score of the top {} pairs of sentences_meta and _sentences_rena: {:.4f}".format(k, average_score))
print("Average cosine similarity score of the top {} pairs of sentences_meta and _sentences_BSH: {:.4f}".format(k, average_score2))
```
