```python
df=data.dropna()
df
```

```python
df = df.reset_index()
df
```

```python
sentences=df['answer']
sentences
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = model.encode(sentences, convert_to_tensor=True)
print(embeddings)

```

```python
print(embeddings.size())
```

```python
!pip install pacmap
import pacmap
```

```python
X=embeddings
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# BASIC
# Initialize the pacmap instance
# Setting n_neighbors to "None" leads to an automatic parameter selection
# choice shown in "parameter" section of the README file.
# Notice that from v0.6.0 on, we rename the n_dims parameter to n_components.
embedding_p = pacmap. PaCMAP (n_components=2, n_neighbors=None, MN_ratio=0.5, FP_ratio=2.0)
# fit the data (The index of transformed data corresponds to the index of the original data)
x_transformed_basic = embedding_p.fit_transform (X, init="pca")
```

```python
x_transformed_basic
```

```python
print(len(x_transformed_basic))
```

```python
x = x_transformed_basic[:, 0]
y = x_transformed_basic[:, 1]
print(len(x))
print(x)

print(len(y))
```

```python
classes = df['AI'].values
print(type(classes))
print(classes)
```

```python
unique_classes = np.unique(classes)
colors = ['r', 'g']
markers = ['d', 'x']

for class_label in unique_classes:
    class_x = x[classes == class_label]
    class_y = y[classes == class_label]
    plt.scatter(class_x, class_y, c=colors[int(class_label)], marker=markers[int(class_label)], label=f'Class {class_label}')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot by Class')
plt.legend()
plt.show()
```
