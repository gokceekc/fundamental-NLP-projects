```python
import pandas as pd
```

```python
data=pd.read_csv('/content/drive/MyDrive/NLP/test.csv')
data
```

```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
sentence=data['answer']
sentence
```

```python
sentence="Taylor Guitars is an American guitar manufacturer based in Richmond, Virginia, that produces a range of instruments under the Guitaria, Guitars and Vibraphones brands. Their most popular models range from custom handmade instruments available to those made upon an order basis. The Guitaria was started in 1989 after a successful business collaboration with one of the leading guitar manufacturers in the world, Guitarsmiths of San Francisco, CA. Their first models were inspired by the music of the 1960s and 1970s, incorporating the style of their favorite groups of the time: The Grateful Dead, The Rolling Stones, Led Zeppelin, The Allman Brothers, The Allman Brothers Band, The All American Rejects, Jan and Dean, and many more."



candidate_labels= ['Human', 'AI']
```

```python
result=classifier(sentence, candidate_labels)
```

```python
result
```

```python
df=pd.read_csv('/content/drive/MyDrive/NLP/train.csv')
df
```

```python
df1 = df.iloc[:10]
```

```python
df1['AI']
```

```python
for i, row in df1.iterrows():
    # concatenate the prompt and answer texts into a single text
    text = str(row['prompt']) + ' ' + str(row['answer'])

    # classify the text using the zero-shot classification pipeline
    result = classifier(text, candidate_labels)

    # extract the top predicted label and its score
    top_label = result['labels'][0]


    # map the predicted label to its numerical value
    predicted_label = 1 if top_label == "AI" else 0

    # store the predicted label in the new column of the dataframe
    df1.at[i, 'predicted_label'] = predicted_label

    # map the actual label to its numerical value
    actual_label = row['AI']
    if actual_label == 0:
        df1.at[i, 'AI'] = 0
    elif actual_label == 1:
        df1.at[i, 'AI'] = 1

print(df1['AI'])
print(df1['predicted_label'])


# calculate the accuracy
accuracy = (df1['AI'] == df1['predicted_label']).mean()
print(f"Accuracy: {accuracy}")
```
